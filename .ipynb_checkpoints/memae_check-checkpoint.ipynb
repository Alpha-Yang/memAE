```python
import torch 
from torch import nn
import summary
import os
import numpy as np
import time
import torch.optim as optim
from torchvision import transforms
from torch.utils.data import DataLoader
import utils
import data.utils as data_utils
import models.loss as loss
from models import AutoEncoderCov3D, AutoEncoderCov3DMem
from tqdm import tqdm
import utils.eval as eval_utils
device = "cuda"
```

```python
import imp
imp.reload(data_utils)
```

```python
height, width = 256, 256
ch = 1
num_frame = 15
batch_size=1
chnum_in_ = 1
mem_dim_in = 2000
sparse_shrink_thres = 0.0025
lr_init = 1e-4
max_epochs = 2
ModelName = "MemAE"
model_path = '/project/bo/exp_data/memory_normal/UCSDped2/original_version_1/model-0059.pt'
# model_path = '/project/bo/exp_data/memory_normal/UCSDped2/original_version_2/model-0059.pt'
# model_path = 'ckpt/MemAE_Conv3DSpar_UCSD_P2_256_MemDim2000_EntW0.0002_ShrThres0.0025_Non.pt'
data_dir = '/project_scratch/bo/anomaly_data/UCSDped2/Train_jpg/'
test_dir = '/project_scratch/bo/anomaly_data/UCSDped2/Test_jpg/'
```

```python
frame_trans = transforms.Compose([
        transforms.Resize([height, width]),
        transforms.Grayscale(num_output_channels=1),
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5]),
    ])
unorm_trans = utils.UnNormalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5))
```

```python
#data_dir = '/project_scratch/bo/anomaly_data/Avenue/frames/training/'
# data_loader = data_utils.DataLoader(data_dir, frame_trans, time_step=num_frame, num_pred=1)
# video_data_loader = DataLoader(data_loader, batch_size=batch_size, shuffle=True, drop_last=True)
data_loader_val = data_utils.DataLoader(test_dir, frame_trans, time_step=num_frame, num_pred=1)
video_data_loader_val = DataLoader(data_loader_val, batch_size=batch_size, shuffle=False, drop_last=True)
```

```python

model = AutoEncoderCov3DMem(chnum_in_, mem_dim_in, shrink_thres=sparse_shrink_thres)
model_para = torch.load(model_path)
model.load_state_dict(model_para)
model.requires_grad_(False)
model.to(device)
model.eval()
```

```python
img_crop_size = 0
recon_error_list = [None] * len(video_data_loader_val)
progress_bar = tqdm(video_data_loader_val)
for batch_idx, frames in enumerate(progress_bar):
    progress_bar.update()
    frames = frames.reshape([batch_size, num_frame + 1, ch, height, width])
    frames = frames.permute(0, 2, 1, 3, 4)
    frames = frames.to(device)
    if (ModelName == 'AE'):
        recon_frames = model(frames)
        ###### calculate reconstruction error (MSE)
        recon_np = utils.vframes2imgs(unorm_trans(recon_frames.data), step=1, batch_idx=0)
        input_np = utils.vframes2imgs(unorm_trans(frames.data), step=1, batch_idx=0)
        r = utils.crop_image(recon_np, img_crop_size) - utils.crop_image(input_np, img_crop_size)
        # recon_error = np.mean(sum(r**2)**0.5)
        recon_error = np.mean(r ** 2)  # **0.5
    elif (ModelName == 'MemAE'):
        recon_res = model(frames)
        recon_frames = recon_res['output']
        recon_np = utils.vframes2imgs(unorm_trans(recon_frames.data), step=1, batch_idx=0)
        input_np = utils.vframes2imgs(unorm_trans(frames.data), step=1, batch_idx=0)
        r = utils.crop_image(recon_np, img_crop_size) - utils.crop_image(input_np, img_crop_size)
        sp_error_map = sum(r ** 2)**0.5
        recon_error = np.mean(sp_error_map.flatten())
    else:
        recon_error = -1
        print('Wrong ModelName.')
    recon_error_list[batch_idx] = recon_error
    if batch_idx >= 0:
        break
```
